```python
maternity = ['infertility']
ayurvedic = ['auyrvedic', 'ayrvedic', 'ayush']
conservative = ['medical management']
opd = ['ophthalmic', 'spectacles', 'panchkarma', 'hearing aids','health checkup', 'chemotherapy oral']
dental = ['denture']
def map_treatment(treatment):
    if pd.isna(treatment):
        return 'unknown'
    
    treatment = str(treatment).lower().strip()

    if treatment in maternity:
        return 'maternity'
    elif treatment in ayurvedic:
        return 'ayurvedic'
    elif treatment in conservative:
        return 'conservative'
    elif treatment in opd:
        return 'opd'
    elif treatment in dental:
        return 'dental'
    else:
        return treatment
```

```python
df['Treatment_type'] = df['Treatment_type'].apply(map_treatment)
```

```python
parents = ['mother', 'father', 'parents', 'uncle', 'aunty', 'guardian', 'grand mother', 'grand father']
in_law = ['daughter in law', 'sister in law', 'brother in law', 'dauil', 'sisil', 'father-in- law', 
          'father-in-law', 'mother in law', 'mother-in-law', 'father in law', 'parent inlaw', 'father inlaw']
childrens = ['son', 'children', 'dependent children', 'ward', 'grand daughter', 'daughter', 'grand son', 'nephew']
self = ['self', 'employee']
cousin = ['sister', 'brother', 'cousin', 'bro', 'sis']
spouse = ['spouse', 'husband', 'wife']
others = ['null', 'nan']

# Map function to categorize the relationship
def map_relationship(relationship):
    if relationship in parents:
        return 'parents'
    elif relationship in in_law:
        return 'in_law'
    elif relationship in childrens:
        return 'children'
    elif relationship in self:
        return 'self'
    elif relationship in cousin:
        return 'cousin'
    elif relationship in spouse:
        return 'spouse'
    elif pd.isna(relationship) or relationship in others:
        return 'others'
    else:
        return 'unknown'

```

```python
df['Relation'] = df['Relation'].apply(map_relationship)
```

```python
x=df.drop(['Opinion'],axis=1)
y=df['Opinion']
```

```python
cat_features = x.select_dtypes(include='object').columns
num_features = x.select_dtypes(exclude='object').columns
```

```python

# Importing necessary transformers
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

# Create transformers for numerical and categorical features
numeric_transformer = StandardScaler()
oh_transformer = OneHotEncoder(drop='first')

# Create the ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ("OneHotEncoder", oh_transformer, cat_features),  # For categorical columns
        ("StandardScaler", numeric_transformer, num_features)  # For numerical columns
    ]
)
```

```python
preprocessor
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=12)
```

```python
x_train=preprocessor.fit_transform(x_train)
```

```python
x_train
```

```python
x_test = preprocessor.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
)
from sklearn.preprocessing import label_binarize
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

# Define your models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "AdaBoost Classifier": AdaBoostClassifier()
}

# Binarize the output for ROC AUC (required for multiclass)
classes = sorted(y_train.unique())
y_train_bin = label_binarize(y_train, classes=classes)
y_test_bin = label_binarize(y_test, classes=classes)

# Imputation step: We'll create an imputer to fill missing values
imputer = SimpleImputer(strategy='mean')  # You can choose 'mean', 'median', or 'most_frequent'

for name, model in models.items():
    # Create a pipeline with imputation and model training
    pipeline = Pipeline(steps=[
        ('imputer', imputer),  # Impute missing values
        ('classifier', model)  # Model
    ])
    
    # Fit the model
    pipeline.fit(x_train, y_train)
    
    # Predictions
    y_train_pred = pipeline.predict(x_train)
    y_test_pred = pipeline.predict(x_test)

    # Probabilities for ROC AUC
    y_train_prob = pipeline.predict_proba(x_train)
    y_test_prob = pipeline.predict_proba(x_test)

    # Train Metrics
    print(f"{name} - Training Set Performance")
    print("- Accuracy: {:.4f}".format(accuracy_score(y_train, y_train_pred)))
    print("- F1 Score: {:.4f}".format(f1_score(y_train, y_train_pred, average='macro')))
    print("- Precision: {:.4f}".format(precision_score(y_train, y_train_pred, average='macro')))
    print("- Recall: {:.4f}".format(recall_score(y_train, y_train_pred, average='macro')))
    print("- ROC AUC Score: {:.4f}".format(roc_auc_score(y_train_bin, y_train_prob, average='macro', multi_class='ovr')))
    print("---------------------------------------------------")

    # Test Metrics
    print(f"{name} - Testing Set Performance")
    print("- Accuracy: {:.4f}".format(accuracy_score(y_test, y_test_pred)))
    print("- F1 Score: {:.4f}".format(f1_score(y_test, y_test_pred, average='macro')))
    print("- Precision: {:.4f}".format(precision_score(y_test, y_test_pred, average='macro')))
    print("- Recall: {:.4f}".format(recall_score(y_test, y_test_pred, average='macro')))
    print("- ROC AUC Score: {:.4f}".format(roc_auc_score(y_test_bin, y_test_prob, average='macro', multi_class='ovr')))
    print("="*50 + "\n")
```
